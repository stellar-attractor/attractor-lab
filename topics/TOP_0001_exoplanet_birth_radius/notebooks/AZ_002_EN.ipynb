# AstroPhysics with Zane


## "Exploring the Sun’s birth radius and the distribution of planet building blocks in the Milky Way galaxy: a multizone Galactic chemical evolution approach" 

(Junichi Baba, Takayuki R. Saitoh and Takuji Tsujimoto, 2023)

https://academic.oup.com/mnras/article/526/4/6088/7324836?login=false

## 1. Purpose of the study
The goal of this paper is to understand where and under what conditions the “building blocks of planets” form in the Milky Way, and what role the Sun occupies in this history.

Unlike studies based on direct exoplanet statistics, this work employs a multizone Galactic Chemical Evolution (GCE) model, which makes it possible to trace:
- how the distribution of chemical elements evolved over time,
- how this evolution affected the availability of materials for planet formation,
- and where in the Galaxy the Sun most likely formed.


## 2. Core idea of the paper
The central idea of the study is that planet formation in the Galaxy is governed not only by stellar age or present-day location, but by the chemical enrichment history of the Galactic disk.

The authors treat the Galaxy as a dynamic, evolving system, in which:
- different radial zones are chemically enriched at different rates,
- the “building blocks of planets” (C, O, Mg, Si, Fe, etc.) become available at different times and locations,
- radial stellar migration mixes this picture into the present-day observed distribution.


## 3. Method: multizone Galactic Chemical Evolution

### 3.1 Multizone model of the Galaxy
The Galactic disk is divided into radial zones, each characterized by:
- its own star formation history,
- its own rate of chemical enrichment,
- its own contributions from supernovae and AGB stars.
The model accounts for:
- gas infall,
- stellar birth and death,
- the release of heavy elements,
- and the gradual flattening of metallicity gradients over time.

### 3.2 Chemical “building blocks of planets”
Special attention is given to elements that are critical for planet formation:
- C and O — volatiles and ices,
- Mg, Si, Fe — silicate and metallic components,
- and the relative abundances of these elements, which determine the types of planets that can form.

The authors emphasize that not only the overall metallicity [Fe/H], but also elemental ratios, are essential.


## 4. The Sun’s birth radius
### 4.1 Reconstructing the birth radius
Using:
- the age of the Sun,
- its chemical composition,
- and the evolution of chemical gradients in the model,
- the authors estimate the Galactocentric birth radius of the Sun, $r_{{birth},\odot}$.

### 4.2 Main result
With high probability, the Sun:
- formed closer to the Galactic center than its current location,
- and subsequently migrated outward.
The estimated values of $r_{{birth},\odot}$ lie inside the present solar radius, not beyond it.
This is consistent with:
- observed chemical peculiarities of the Sun,
- independent estimates of radial migration.


## 5. Distribution of planetary building blocks in the Galaxy
### 5.1 Inner Galaxy
In the inner regions of the Galactic disk:
- gas densities are higher,
- star formation is more intense,
- heavy elements accumulate earlier and more rapidly.
As a result:
- planetary building blocks become available earlier,
- the probability of forming massive planets is higher.

### 5.2 Outer Galaxy
At larger Galactocentric distances:
- chemical enrichment proceeds more slowly,
- heavy elements are scarce for extended periods,
- planet formation begins later and is less efficient.

Over time, conditions improve — but with a significant delay.


## 6. The role of radial migration
The authors stress that:
- A star’s present-day position does not reflect the conditions under which its planets formed.

Radial migration:
- transports stars from the inner Galaxy to the outer disk,
- creates the illusion of “planets existing where they could not have formed,”
- explains why planetary systems are observed at radii with low present-day metallicity.


## 7. Connection to observed exoplanets
This work provides the theoretical foundation for the results of
“Where in the Milky Way do exoplanets preferentially form?”:
- both studies conclude that planets preferentially form closer to the Galactic center;
- differences between massive and low-mass planet populations naturally follow from chemical evolution;
- the birth radius is a key parameter without which observational interpretations remain incomplete.

## 8. Model limitations
The authors explicitly acknowledge that:
- the model depends on assumptions about star formation rates;
- uncertainties exist in supernova and AGB stellar yields;
- radial migration is parameterized rather than modeled on individual orbits.

Nevertheless, the qualitative conclusions are robust.

## 9. What is especially valuable for visualization and content
For future work (Python, Academy, video production):
- time-dependent radial metallicity profiles;
- evolution of the availability of C, O, Mg, Si, Fe;
- comparison of the inner versus outer Galaxy;
- reconstruction of the Sun’s path from birth to its present position.


## 10. Short conclusion

- Planet formation is a product of Galactic chemical history.
- The Sun most likely formed closer to the Galactic center.
- Radial migration is critical for understanding exoplanet populations.
- The present-day map of planets is a mixed projection of the past.
