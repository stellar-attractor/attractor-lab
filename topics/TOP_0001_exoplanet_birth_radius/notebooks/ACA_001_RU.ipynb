# Академия: Модуль 1 

## Как восстановить карту рождения планет в Млечном Пути

## 1. Зачем вообще понадобились эти понятия

*За последние десятилетия мы поняли, что планеты — это не экзотика, а обычный продукт звездообразования. Но привычный вопрос «есть ли планеты?» сменился более фундаментальным: где именно в Галактике они формируются и почему мы видим их именно там, где видим сегодня. Чтобы ответить на него, нам приходится выйти за рамки локальной астрономии и посмотреть на планетообразование в галактическом контексте.*

При анализе экзопланетных систем быстро выяснилось противоречие:
планеты обнаруживаются у звёзд, находящихся сегодня в областях Галактики, где условия для их формирования выглядят неблагоприятными (низкая металличность, малое количество строительных блоков).

Это противоречие исчезает, если отказаться от предположения, что звезда родилась там же, где мы её наблюдаем сейчас.

Так возникли два ключевых понятия:
- Birth Radius ($r_{birth}$) — где звезда родилась;
- Stellar Migration — почему она находится в другом месте.

## 2. Birth Radius ($r_{birth}$): формальное определение
*Очень важно сразу подчеркнуть: формирование планет нельзя рассматривать как локальное свойство конкретной звезды или её окрестностей. Оно отражает историю всей Галактики — то, как она накапливала тяжёлые элементы и как со временем перемещались сами звёзды. Поэтому наблюдаемое распределение экзопланет — это не карта мест их рождения, а результат длительной эволюции.*

Birth radius ($r_{birth}$) — это галактоцентрическое расстояние, на котором звезда сформировалась, восстановленное на основе:
- возраста звезды,
- её химического состава,
- модели химической эволюции Галактики.

Важно:
- $r_{birth}$ не является наблюдаемой величиной;
- это реконструкция, основанная на физически мотивированных моделях.


## 3. Физическая основа $r_{birth}$

*Здесь появляется ключевое понятие — радиус рождения. Звёзды не остаются на тех же орбитах, где сформировались, поэтому их текущее положение часто вводит в заблуждение. Радиус рождения позволяет восстановить исходную среду формирования звезды и её планет, используя возраст, металличность и модели химической эволюции Галактики.*

### 3.1 Градиент металличности
В диске Млечного Пути наблюдается:
- отрицательный радиальный градиент металличности;
- внутренняя Галактика богаче тяжёлыми элементами, чем внешняя.

### 3.2 Эволюция градиента со временем
Градиент:
- был более крутым в прошлом;
- со временем выравнивался за счёт звездообразования и перемешивания газа.

Следствие:
- одинаковая металличность в разные эпохи соответствует разным радиусам.

## 4. Как практически оценивается $r_{birth}$
Типовая процедура:
- Определяется возраст звезды (изохроны, спектроскопия, астеросейсмология).
- Измеряется химический состав ($[Fe/H]$, $\alpha$-элементы).
- Используется модель Galactic Chemical Evolution: $[Fe/H](R, t)$.
- Находится радиус $R$, при котором наблюдаемая химия возможна в момент рождения.

Результат:
распределение вероятностей $r_{birth}$, а не одна точка.


## 5. Основные механизмы миграции звезд

Даже если $r_{birth}$ восстановлен корректно, остаётся вопрос: почему звезда не находится там сейчас? Ответ — миграция звезд. 

### 5.1 Орбитальное размывание
- эксцентрические орбиты;
- радиальные колебания вокруг среднего радиуса;
- средний радиус не меняется.

Орбитальное размывание создаёт разброс, но не меняет $r_{birth}$.


### 5.2 Радиальное перемещение
- резонансные взаимодействия со спиральными рукавами и перемычкой Галактики;
- изменение углового момента;
- переход на почти круговую орбиту с другим радиусом.

Радиальное перемещение — главный механизм долгосрочной миграции.


## 6. Связь звездной миграции и экзопланет
Планеты:
- формируются в первые ~10–100 млн лет;
- фиксируют химические условия места рождения.

Если звезда затем мигрирует:
- планетная система “переезжает” целиком;
- химия остаётся неизменной.

Это объясняет:
- наличие планет у звёзд во внешнем диске;
- кажущиеся “аномалии” металличности.


## 7. Солнце как контрольный пример
Модели показывают, что:
- металличность Солнца выше средней для его текущего расстояния от центра Галактики;
- наиболее вероятный $r_{birth}$ Солнца ($6.5\pm0.1$ кпк) меньше текущего расстояния от центра Галактики ($8.2\pm0.1$ кпк).

Следствие:
- Солнце — мигрант;
- Земля сформировалась из вещества, типичного для более внутренней Галактики.


## 8. Почему без этих понятий нельзя интерпретировать экзопланеты

Без понятий $r_{birth}$ и звездной миграции:
- искажается статистика экзопланет;
- смешиваются внутренние и внешние области Галактики;
- теряется связь с химической эволюцией.

С ними:
- появляется физически связная картина;
- экзопланеты становятся частью истории Галактики, а не локальными исключениями.


## 9. Заключение

- Планеты формируются в конкретных галактических условиях.
- Эти условия зависят от радиуса и эпохи.
- Звёзды мигрируют, но их химическая память сохраняется.
- $r_{birth}$ и миграция звезд — ключ к реконструкции истории планет.
