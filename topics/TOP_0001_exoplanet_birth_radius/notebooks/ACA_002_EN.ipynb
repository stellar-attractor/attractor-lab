# Academy Module 2 Galactic Chemical Evolution (GCE): How the Milky Way Gradually Created the Elements for Planets

## 1. What Galactic Chemical Evolution is
Galactic Chemical Evolution (GCE) is the set of physical processes that describe how the chemical composition of the Galaxy has changed over time and as a function of Galactocentric radius.

Put simply:
GCE answers the question of when, where, and from what planet formation became possible in the Galaxy.

GCE is not an abstract theory, but a framework that brings together:
- star formation,
- nuclear synthesis in stars,
- the release of elements into the interstellar medium,
- the redistribution of gas and stars.


## 2. Why planets cannot be discussed without GCE
Planets do not form “out of nothing.”
They form from interstellar gas whose chemical composition:
- evolved over time,
- varied with Galactocentric radius.

As a result:
- the early Galaxy → few heavy elements → few planets;
- the inner Galaxy → faster enrichment → more planetary building blocks;
- the outer Galaxy → delayed chemical development.

Without GCE, it is impossible to understand:
- why massive planets are rare in old stellar populations;
- why planet formation “turned on” at different times at different radii.


## 3. Main sources of chemical elements
### 3.1 Hydrogen and helium
formed in the first minutes after the Big Bang;
make up most of the gas mass, but do not build solid planets.

### 3.2 Massive stars and Type II supernovae
evolve rapidly (millions of years);
produce:
- oxygen,
- magnesium,
- silicon,
- other α-elements.
These are the first major suppliers of planetary building materials.

### 3.3 Type Ia supernovae
occur later (hundreds of millions to billions of years);
the main source of iron.
They:
- gradually increase [Fe/H];
- make the formation of massive planets possible.

### 3.4 AGB stars
enrich the interstellar medium with:
- carbon,
- nitrogen,
- s-process elements;
play an important role in the chemistry of ices and organic compounds.


## 4. Why chemistry varies with Galactic radius
### 4.1 The inner disk
- high gas density;
- intense star formation;
- rapid cycle of “birth → death → enrichment.”

Result:
high metallicity already at early epochs.

### 4.2 The outer disk
- low gas density;
- slow star formation;
- delayed enrichment.

Result:
a deficit of planetary building blocks in the early Galaxy.


## 5. The metallicity gradient
A robust observational trend is observed:
- [Fe/H] decreases with increasing Galactocentric radius.

Importantly:
- this gradient is not constant;
- it flattens over time due to:
- ongoing star formation,
- gas infall,
- radial mixing.


## 6. Multizone GCE: why a single zone is not enough
In multizone models:
- the Galaxy is divided into radial rings;
- each ring has:
- its own star formation history,
- its own chemical evolution.

This approach allows one to:
- track where and when elements appear;
- correctly introduce the birth radius;
- link chemistry with stellar migration.

This is precisely the approach used in the second paper.


## 7. Planetary building blocks in the context of GCE
For planets, the most important elements are:
- Fe — planetary cores and massive planets;
- Mg, Si — silicate mantles;
- O, C — ices, water, organic chemistry;
- C/O — atmospheric types and mineralogy.

GCE determines:
- when these elements became available;
- where they existed in the required proportions.


## 8. The connection between GCE, birth radius, and migration
GCE defines the map:
- [elements] = f(R, t).

The birth radius tells us which point on this map the star “remembers”;
stellar migration explains why the star is now located elsewhere.

Together:
- GCE = the stage
- r₍birth₎ = the starting point
- migration = the trajectory


## 9. Why GCE makes the exoplanet picture coherent
When GCE is taken into account:
- massive planets are more common around stars of inner-Galaxy origin;
- low-mass planets are possible almost everywhere, but at later times;
- the distribution of exoplanets becomes a historical narrative rather than a collection of anomalies.


## 10. What should be visualized in the Academy
A minimal recommended set:
- evolution of [Fe/H] at different epochs;
- radial evolution of Fe, Mg, Si, and O;
- zones of “availability” of planetary building blocks;
- schematic: GCE → r₍birth₎ → migration → observed system.


## 11. Core takeaway of the module
- The Galaxy was not always capable of forming planets.
- It learned to do so as chemical enrichment progressed.
- Planets are a product not only of stars, but of the evolution of the entire Galaxy.
