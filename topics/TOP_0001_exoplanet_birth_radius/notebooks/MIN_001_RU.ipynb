# Минутка астрофизики: Где в Млечном Пути рождаются планеты?

Мы видим экзопланеты по всей Галактике.

Но это не значит, что они там родились.

Звёзды умеют мигрировать.

За миллиарды лет они меняют свои орбиты и могут уехать на тысячи световых лет от места своего рождения.

Поэтому астрономы используют понятие радиус рождения звезды — birth radius.

Это то место в Галактике, где звезда сформировалась, и где вокруг неё могли появиться планеты.

Внутренние области Млечного Пути раньше и быстрее обогащались тяжёлыми элементами.

А именно из них строятся планеты.

Когда учёные восстановили радиусы рождения, оказалось:

- планеты чаще рождаются ближе к центру Галактики,
- а затем путешествуют вместе со своими звёздами.

Даже наше Солнце, вероятно, родилось ближе к центру Млечного Пути.

Планеты — не просто точки на карте.

Это путешественники, несущие химическую память Галактики.
