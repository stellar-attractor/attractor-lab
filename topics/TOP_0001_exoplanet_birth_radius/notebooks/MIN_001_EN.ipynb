# 1 minute of astrohysics — Where in the Galaxy do planets form?

We see exoplanets all across the Galaxy.

But that doesn’t mean they were born there.

Stars can migrate.

Over billions of years, they change their orbits
and can travel thousands of light-years away from their birthplace.

That’s why astronomers use the concept of a star’s birth radius.

The birth radius is the location in the Galaxy where a star formed
— and where planets could first emerge around it.

The inner regions of the Milky Way
were enriched with heavy elements earlier and more rapidly.

And it is from these elements that planets are built.

When scientists reconstructed stellar birth radii, they found that:

	•	planets are more likely to form closer to the Galactic center,
	•	and later travel through the Galaxy together with their stars.

Even our Sun was likely born closer to the center of the Milky Way.

Planets are not just points on a map.

They are travelers,
carrying the Galaxy’s chemical memory with them.
