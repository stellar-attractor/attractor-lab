# Академия · Модуль 2 

## Химическая эволюция Галактики: как Млечный Путь постепенно создавал элементы для планет

## 1. Что такое Химическая эволюция Галактики (GCE)

Химическая эволюция Галактики (GCE) — это совокупность физических процессов, которые описывают, как химический состав Галактики менялся со временем и по радиусу.

Проще:
GCE отвечает на вопрос, когда, где и из чего в Галактике стало возможно формирование планет.

GCE — это не абстрактная теория, а рамка, которая объединяет:
- звездообразование,
- ядерный синтез в звёздах,
- выбросы элементов в межзвёздную среду,
- перераспределение газа и звёзд.


## 2. Почему без GCE нельзя говорить о планетах
Планеты не возникают “из ничего”.

Они формируются из межзвёздного газа, химический состав которого:
- менялся со временем;
- различался по галактическому радиусу.

Следовательно:
- ранняя Галактика → мало тяжёлых элементов → мало планет;
- внутренняя Галактика → быстрее обогащение → больше строительных блоков;
- внешняя Галактика → задержка в химическом развитии.

Без GCE невозможно понять:
- почему массивные планеты редки в старых популяциях;
- почему планетообразование “включалось” в разное время на разных радиусах.


## 3. Основные источники химических элементов

### 3.1 Водород и гелий
образовались в первые минуты после Большого взрыва;
составляют основную массу газа, но не строят твёрдые планеты.

### 3.2 Массовые звёзды и сверхновые II типа
быстро эволюционируют (миллионы лет);
производят: кислород, магний, кремний, другие α-элементы.
Это первые “поставщики” строительных материалов.

### 3.3 Сверхновые Ia
происходят позже (сотни миллионов – миллиарды лет);
основной источник железа.
Именно они постепенно повышают [Fe/H] и делают возможным формирование массивных планет.

### 3.4 AGB-звёзды
обогащают среду углеродом, азотом, s-элементами; важны для химии льдов и органики.


## 4. Почему химия меняется по радиусу Галактики
### 4.1 Внутренний диск
- высокая плотность газа;
- интенсивное звездообразование;
- быстрый цикл “рождение → смерть → обогащение”.

Результат: высокая металличность уже на ранних этапах.

### 4.2 Внешний диск
- низкая плотность газа;
- медленное звездообразование;
- запаздывающее обогащение.

Результат: дефицит строительных блоков планет в ранние эпохи.


## 5. Градиент металличности

Наблюдается устойчивая закономерность:
$[Fe/H]$ уменьшается с увеличением галактоцентрического радиуса.

Важно: этот градиент не постоянен; со временем он уплощается из-за:
- звездообразования,
притока газа,
радиального перемешивания.


## 6. Многозонная GCE: почему одной зоны недостаточно

В многозонных моделях:
- Галактика разбивается на радиальные кольца;
- каждое кольцо имеет:
- собственную историю звездообразования;
- собственную химическую эволюцию.

Это позволяет:
- отслеживать, где и когда появляются элементы;
- корректно вводить радиусы рождения;
- связать химию с миграцией.


## 7. Строительные блоки планет в контексте GCE
Для планет особенно важны:
- Fe — ядра и массивные планеты;
- Mg, Si — силикатные мантии;
- O, C — лёд, вода, органика;
- C/O — типы планетных атмосфер и минералогия.

GCE определяет:
- когда эти элементы появились;
- где они были доступны в нужных пропорциях.


## 8. Связь GCE с радиусом рождения и звездной миграцией

GCE задаёт карту: $[элементы] = f(R, t)$.

Радиус рождения позволяет определить, какую точку карты “запомнила” звезда.
stellar migration объясняет, почему звезда сейчас находится в другой точке.

Вместе:
- GCE = сцена
- r_birth = точка старта
- migration = траектория


## 9. Почему GCE делает картину экзопланет логичной
С учётом GCE:
- массивные планеты чаще у звёзд внутреннего происхождения;
- маломассивные планеты возможны почти везде, но позже;
- распределение экзопланет становится историей, а не набором исключений.

## 12. Вывод
Галактика не всегда могла создавать планеты.
Она училась делать это по мере химического обогащения.
Планеты — продукт не только звёзд, но и эволюции всей Галактики.
