# Academy · Module 1 How to Reconstruct the Birth Map of Planets in the Milky Way

## 1. Why these concepts were needed in the first place
When analyzing exoplanetary systems, a contradiction quickly became apparent:
planets are detected around stars that are currently located in regions of the Galaxy where conditions for planet formation appear unfavorable (low metallicity, a scarcity of planetary building blocks).

This contradiction disappears if we abandon the assumption that:
a star was born where we observe it today.

This led to the introduction of two key concepts:
- Birth Radius ($r_{birth}$) — where a star was born;
- Stellar Migration — why it is found elsewhere today.

## 2. Birth Radius ( $r_{birth}$): formal definition
The birth radius ($r_{birth}$) is the Galactocentric distance at which a star formed, reconstructed using:
- the star’s age,
- its chemical composition,
- a model of Galactic chemical evolution.

Importantly:
$r_{birth}$ is not an observable quantity, it is a reconstruction based on physically motivated models.

## 3. Physical basis of  $r_{birth}$
### 3.1 Metallicity gradient
In the Milky Way disk, we observe:
- a negative radial metallicity gradient;
- the inner Galaxy is richer in heavy elements than the outer regions.

### 3.2 Time evolution of the gradient
The gradient:
- was steeper in the past;
- gradually flattened due to star formation and gas mixing.

### 3.3 Consequence:
*The same metallicity corresponds to different Galactocentric radii at different epochs.*

## 4. How $r_{birth}$ is estimated in practice
A typical procedure:

1. The stellar age is determined (isochrones, spectroscopy, asteroseismology).

2. The chemical composition is measured ([Fe/H], α-elements).

3. A Galactic Chemical Evolution model is used:
$[\mathrm{Fe/H}](R, t)$.

4. The radius R is identified where the observed chemistry is possible at the star’s birth time.

Result:
A probability distribution of $r_{birth}$, not a single point.


## 5. Stellar migration: why it is necessary
Even if $r_{birth}$ is correctly reconstructed, a key question remains:

Why is the star not located there today?

The answer is stellar migration.

## 6. Main mechanisms of stellar migration

### 6.1 Blurring

- eccentric stellar orbits;
- radial oscillations around a mean radius;
- the mean radius does not change.

*Blurring introduces scatter but does not change $r_{birth}$.*

### 6.2 Churning

- resonant interactions with spiral arms and the Galactic bar;
- changes in angular momentum;
transition to a nearly circular orbit at a different radius.

*Churning is the primary mechanism of long-term radial migration.*

## 7. The connection between migration and exoplanets
Planets:
- form within the first ~10–100 million years;
- record the chemical conditions of their birth environment.

If the star later migrates:
- the entire planetary system migrates with it;
- the chemical composition remains unchanged.

This explains:
- the presence of planets around stars in the outer disk;
- apparent metallicity “anomalies.”


## 8. The Sun as a reference case
Models show that:
- the Sun’s metallicity is higher than the average at its current radius;
- the most probable r₍birth₎ of the Sun is smaller than its present Galactocentric distance.

*Implications:*
- the Sun is a migrant;
- Earth formed from material typical of the inner Galaxy.


## 9. Why exoplanets cannot be interpreted without these concepts
Without $r_{birth}$ and migration:
- exoplanet statistics are distorted;
- inner and outer Galactic regions are mixed;
- the link to chemical evolution is lost.

With them:
- a physically coherent picture emerges;
- exoplanets become part of the Galaxy’s history, not local exceptions.

## 11. Core takeaway of the module
1. Planets form under specific Galactic conditions.
2. These conditions depend on radius and epoch.
3. Stars migrate, but their chemical memory remains.
4. Birth radius and stellar migration are the key to reconstructing the history of planets.
