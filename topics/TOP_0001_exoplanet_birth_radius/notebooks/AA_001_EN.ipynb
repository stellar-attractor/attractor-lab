# Where Planets Preferentially Form in the Milky Way and Why We Observe Them Far from Their Birthplace

## 1. Introduction: planets as a Galactic phenomenon

The discovery of thousands of exoplanets has transformed planets from rare curiosities into a fundamental outcome of star formation. Yet, despite the growing census, a central question remains unresolved: where in the Galaxy do planets preferentially form, and why are they observed where they are today?

Two recent lines of research converge on a common answer. Planet formation is not a purely local process tied only to the present-day environment of a star. Instead, it reflects the long-term chemical evolution of the Milky Way combined with stellar orbital migration. The observed distribution of exoplanets is therefore not a direct map of planet formation sites, but a blurred snapshot of a much older Galactic history.

## 2. Birth radius as the missing coordinate

A key concept linking these studies is the birth radius, $r_{\rm birth}$: the Galactocentric distance at which a star (and its planetary system) originally formed.

In exoplanet studies, $r_{\rm birth}$ is reconstructed by combining stellar ages with metallicities and models of Galactic chemical evolution. This approach acknowledges that a star’s current position is often misleading: stars do not remain fixed at their formation radius over billions of years.

Independent work on the chemical origins of planetary building blocks demonstrates why this reconstruction is physically meaningful. A star’s atmospheric composition encodes the metallicity of the interstellar medium at the time and place of its birth. While stellar migration redistributes stars across the Galactic disk, their chemistry acts as a fossil record, preserving information about their origin.

The implication is clear:
to understand where worlds are born, we must look beyond where stars are observed today and instead reconstruct where they formed.

## 3. The inner Galaxy as a factory of worlds

Both studies support a consistent causal narrative rooted in Galactic chemical evolution.

The Milky Way did not enrich itself uniformly. The inner disk experienced:
- higher star formation rates,
- more frequent supernova events,
- faster enrichment in heavy elements.

As a result, the raw materials required for planet formation—iron, silicon, magnesium, carbon, oxygen—became available earlier and in greater abundance closer to the Galactic center. In contrast, the outer disk lagged behind, reaching comparable chemical maturity only later in cosmic time.

This naturally leads to a robust conclusion:
planet formation efficiency decreases with Galactocentric distance, and this trend is especially pronounced for high-mass (giant) planets, which require both metal-rich environments and sufficiently massive protoplanetary disks.

In this framework, the inner Galaxy functioned as an early and efficient “factory of worlds,” while the outer disk became hospitable to planet formation only after prolonged chemical enrichment.

## 4. Why planets are found in “the wrong places”

If planets form more efficiently in the inner Galaxy, why do we observe many planetary systems at large Galactocentric radii today?

The answer lies in stellar radial migration. Over billions of years, stars can exchange angular momentum with spiral arms and Galactic structures, drifting inward or outward by several kiloparsecs. A planetary system may therefore be born in the chemically rich inner disk and later be observed far from its birthplace.

This process provides an intuitive explanation that requires no equations:

The Galaxy is not a static map with pins—it is an ocean of currents.
Stars are not fixed points; they are travelers.

Planetary systems move with their host stars, carrying the imprint of their birth environments across the disk.

## 5. The Sun as a narrative anchor

The Sun offers a powerful human-scale illustration of this idea. Independent chemical and dynamical evidence suggests that the Sun likely formed closer to the Galactic center and later migrated outward to its present location.

This reframes our perspective on Earth itself. The material from which our planet formed did not originate “here” in the present-day solar neighborhood, but in a region of the Galaxy that was chemically richer at an earlier epoch.

In this sense:
- our home is a migrant,
- and Earth is built from matter whose story began elsewhere in the Milky Way.

This narrative fits seamlessly with the broader exoplanet picture: if planet formation is more efficient in the inner disk, and migration redistributes stars outward, then the current Galactic geography of planets is shaped as much by motion as by birth.

## 6. A unified picture

Taken together, these studies converge on a single framework.

The Milky Way first developed its planet-forming potential in the inner regions of its disk, where chemical enrichment proceeded rapidly. Over time, as heavy elements accumulated and metallicity gradients flattened, the zone capable of forming planets expanded outward. On average, planets—especially massive ones—formed more efficiently closer to the Galactic center.

However, stellar migration subsequently redistributed planetary systems across the disk. As a result, the present-day locations of exoplanets do not directly trace their formation sites. To reconstruct where worlds are truly born, one must introduce the concept of the birth radius and account for Galactic chemical evolution and stellar migration together.

## 7. Methodological reconstruction (how it works)

Physically, stellar metallicity is determined by the chemical composition of the interstellar medium at the time and place of star formation. In this sense, metallicity is a function of Galactocentric radius and time:

$$ [Fe/H](r,t) = f(r_{birth},t) $$

Observed $[Fe/H]$ + stellar age> Galactic chemical evolution model > Reconstructed $r_{birth}$

However, to reconstruct the $r_{birth}$, scientists invert the physical relation. Using the observed metallicity and the stellar age, together with a model of Galactic chemical evolution, we reconstruct the Galactocentric radius at which such chemical conditions were typical at the time of the star’s formation.

Formally, this corresponds to:

$$ r_{birth} = f^{-1}([Fe/H],\text{age}) $$

Metallicity is not treated as a causal variable in this analysis, but as a tracer of the stellar birth environment. Thus, figures showing $r_{birth}$ as a function of $[\mathrm{Fe/H}]$ should be interpreted as consistency checks and reconstructions, not as statements about physical causality running from metallicity to radius.

To assess whether the results obtained with the toy model remain valid under physically realistic conditions, we compare them with a multizone chemo-dynamical model of Galactic Chemical Evolution (GCE). In this framework, the Milky Way disk is divided into radial annuli, each characterized by its own star formation history and chemical enrichment rate. The model explicitly accounts for the main nucleosynthetic sources — core-collapse supernovae (Type II), Type Ia supernovae, and AGB stars — and follows the time-dependent evolution of elemental abundances across the disk. Crucially, stellar radial migration is included through both orbital blurring and churning, allowing stars to change their mean Galactocentric radii while retaining the chemical signatures of their birth environments. This produces a realistic spread of stellar metallicities at a given present-day radius and naturally motivates the introduction of the birth radius, r_{\mathrm{birth}}.

The structure and philosophy of this model closely follow the chemo-dynamical models developed by Minchev et al. (e.g. 2013, 2014, 2018), which combine Galactic dynamics with chemical evolution in a self-consistent way. These models successfully reproduce key observational constraints, including the present-day metallicity gradient, its temporal evolution, the dispersion in [Fe/H] and α-element abundances, and trends observed in large surveys such as APOGEE, Gaia, and HARPS. By comparing the toy model against this Minchev-type multizone GCE framework, we demonstrate that the inferred trends in planet formation and birth radius are not artifacts of simplified assumptions, but instead emerge robustly within a physically motivated description of the Milky Way’s chemo-dynamical evolution.
