# Where Planets Form in the Milky Way

## 1. Planets as a Galactic phenomenon

The discovery of thousands of exoplanets has shown that planets are a natural outcome of star formation. Yet a fundamental question remains: where in the Milky Way do planets form most efficiently, and why are they observed where they are today?

Recent studies converge on a common picture. Planet formation is not a purely local process tied to a star’s present-day environment. Instead, it reflects the long-term chemical evolution of the Galaxy combined with stellar migration. The observed distribution of exoplanets is therefore not a direct map of their birthplaces, but a blurred imprint of Galactic history.

## 2. Birth radius as a key concept

A central quantity is the birth radius, $r_{\rm birth}$ — the Galactocentric distance at which a star and its planetary system originally formed.

Because stars migrate over billions of years, their current locations are often poor indicators of their formation environments. The birth radius can be reconstructed by combining stellar ages, metallicities, and models of Galactic chemical evolution.

Stellar chemistry provides the physical basis for this approach. The atmospheric composition of a star preserves the metallicity of the interstellar medium at the time and place of its formation, acting as a fossil record of its birth environment.

## 3. The inner Galaxy as an efficient planet-forming region

The Milky Way enriched itself chemically in a highly non-uniform way. The inner disk experienced higher star formation rates, more frequent supernovae, and faster enrichment in heavy elements.

As a result, the material required for planet formation became available earlier and in greater abundance closer to the Galactic center. The outer disk reached comparable chemical maturity only later. Consequently, planet formation efficiency decreases with Galactocentric radius, an effect that is especially strong for giant planets.

## 4. Why planets are found far from their birthplaces

If planets form preferentially in the inner Galaxy, why do we observe many planetary systems at large Galactocentric radii?

The answer lies in stellar radial migration. Interactions with spiral arms and other Galactic structures allow stars to move inward or outward by several kiloparsecs over time. Planetary systems migrate together with their host stars, carrying the chemical imprint of their birth environments across the disk.

## 5. The Sun as an illustration

The Sun provides a familiar example. Chemical and dynamical evidence suggests that it likely formed closer to the Galactic center and later migrated outward to its present location.

This implies that the material from which Earth formed originated in a region of the Galaxy that was chemically richer at an earlier epoch.

## 6. A unified picture

Together, these results point to a coherent framework. Planet formation was initially most efficient in the inner Milky Way, where chemical enrichment proceeded rapidly, and later expanded outward as the disk evolved. Stellar migration subsequently redistributed planetary systems, breaking the direct link between present-day location and formation site.

## 7. Methodological robustness

Physically, stellar metallicity is determined by the chemical composition of the interstellar medium at the time and place of star formation. In this sense, metallicity is a function of Galactocentric radius and time:

$$ [Fe/H](r,t) = f(r_{birth},t) $$

Observed $[Fe/H]$ + stellar age> Galactic chemical evolution model > Reconstructed $r_{birth}$

However, to reconstruct the $r_{birth}$, scientists invert the physical relation. Using the observed metallicity and the stellar age, together with a model of Galactic chemical evolution, we reconstruct the Galactocentric radius at which such chemical conditions were typical at the time of the star’s formation:

$$
r_{\mathrm{birth}} = f^{-1}([\mathrm{Fe/H}], \mathrm{age})
$$

Here, metallicity is used as a tracer of the birth environment rather than as a causal variable.

To test robustness, results obtained with a simplified (“toy”) model are compared with more realistic chemo-dynamical models that include detailed enrichment histories and stellar migration.

## 8. Conclusion

Exoplanets in the Milky Way should be understood not as a static map of formation sites, but as the outcome of a dynamically evolving Galaxy. To understand where worlds are truly born, one must consider both the chemical history of the Milky Way and the long-term migration of its stars.
