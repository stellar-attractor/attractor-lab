## “Where in the Milky Way do exoplanets preferentially form?”

(Teixeira, Adibekyan, Bossini et al., 2025)

https://arxiv.org/pdf/2501.11660


## 1. Purpose of the study
Exoplanets are detected around stars that are currently located in very different regions of the Milky Way. However, stars migrate over time, so their present-day position does not reflect the conditions under which planet formation took place.

The goal of this work is to determine the Galactic birth radii of stars and their planets (birth radius, r₍birth₎) and, on this basis, to understand where in the Galaxy planet formation was most efficient and how this efficiency evolved over time.


## 2. Main idea of the paper
The key idea of the study is that planet formation is part of Galactic chemical evolution, rather than a purely local process determined only by individual stellar properties.

The authors reconstruct the birth locations of stars hosting planets using their ages and metallicities and show that:
- planetary systems, on average, formed closer to the Galactic center;
- the efficiency of planet formation decreases with increasing Galactocentric distance;
- this effect is particularly strong for massive planets.


## 3. Data and methodology
### 3.1 Data samples
The study is based on two independent datasets:

**SWEET-Cat** 
- a large compilation of stars with detected exoplanets;
- used to derive statistically robust distributions.

**HARPS-GTO** 
- a high-quality spectroscopic sample including:
- stars hosting planets,
- a control sample of stars without detected planets.

This duplication allows the authors to verify that the results are not artifacts of a specific catalog.

### 3.2 Stellar parameter estimation
For each star, the following parameters are determined:
- Age — based on photometric, spectroscopic, and astrometric data;
- Metallicity ([Fe/H]) — serving as an indicator of the chemical composition of the gas from which the star formed. 

### 3.3 Reconstruction of the birth radius ($r_{birth}$)
The stellar birth radius is reconstructed in a model-dependent way by:
- using the stellar age;
- matching the star’s metallicity to models of Galactic chemical evolution;
- identifying the Galactocentric radius at which such chemical conditions were typical at the time of the star’s formation.

Thus, $r_{birth}$ is not a directly observed quantity, but a physically motivated reconstruction.

## 4. Key results
### 4.1 Metallicity
Stars hosting detected planets are, on average:
- more metal-rich
- than stars without detected planets.

This result is consistent with the idea that the presence of heavy elements enhances the efficiency of planet formation, particularly for massive planets.

### 4.2 Age
Planet-hosting stars are, on average:
- younger
- than stars without planets.
This reflects both genuine astrophysical effects and observational limitations, as young active stars are more difficult to survey for planets.

### 4.3 Birth radius
The central result of the paper is that:
Stars hosting planets have smaller birth radii ($r_{birth}$) than stars without detected planets.
On average:
- planet-hosting stars were born closer to the Galactic center;
- stars without planets formed at larger Galactocentric distances.
This implies that planet formation was more efficient in the inner regions of the Galactic disk.

### 4.4 Dependence on planet mass
The authors divide planetary systems into:
- systems with massive planets;
- systems with low-mass planets.

They find that:
- planet formation efficiency decreases with increasing $r_{birth}$;
- this decrease is significantly stronger for massive planets than for low-mass ones.

This indicates that the formation of gas giants is particularly sensitive to the Galactic environment.


## 5. Evolution over time
By considering different stellar age intervals, the authors show that:
- in the early epochs of the Galaxy, planet formation efficiency was lower;
- over time, as the disk became chemically enriched, the region of efficient planet formation expanded outward;
- this outward expansion is more pronounced for massive planets than for low-mass ones.

Thus, the Milky Way became progressively more favorable for planet formation as it evolved.


## 6. Interpretation of the results
The results are naturally explained by a combination of:
- the negative metallicity gradient of the interstellar medium;
- the gradual chemical enrichment of the Galaxy;
- radial stellar migration.

Exoplanet formation emerges as a global Galactic process, closely tied to the overall evolution of the disk.

## 7. Limitations and observational biases
The authors emphasize several important caveats:
- low-mass planets are harder to detect around young, active stars;
- some differences may be amplified by observational biases;
- the birth radius is a model-dependent quantity and depends on the adopted chemical evolution model.

Nevertheless, the main trends are robust and are reproduced across independent datasets.


## 8. What can be visualized
For further work (Academy content, Python plots), particularly useful diagnostics include:
- distributions of [Fe/H], age, and $r_{birth}$ for:
    - stars with planets,
    - stars without planets;
- the dependence of planet formation efficiency on $r_{birth}$;
- comparisons between massive and low-mass planets;
- the evolution of these relations with stellar age.


## 9. Short conclusions
- Planet formation is not uniform across the Galactic disk.
- The inner regions of the Milky Way were more favorable for planet formation.
- Massive planets are especially sensitive to the Galactic chemical environment.
- The history of exoplanets is tightly linked to the chemical and dynamical evolution of the Galaxy.
